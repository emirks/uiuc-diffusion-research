# exp_024 — LTX-2 Prompt Optimization Sweep — run_0003

| Cat | What it tests |
|-----|---------------|
| **A-empty** | No prompt (model floor) |
| **A-word** | Single word: "transition" |
| **B** | Generic cinematic structure |
| **C** | Abstract object-agnostic descriptor |
| **D** | Named mechanism + explicit objects |
| **E** | Temporal language (at 4s / by 6s) |


```python
import pathlib, yaml
from IPython.display import display, Video, Markdown

REPO_ROOT   = pathlib.Path('..').resolve()
RUN_DIR     = REPO_ROOT / 'outputs/videos/exp_024_ltx2_prompt_sweep/run_0003'
CONFIG_PATH = REPO_ROOT / 'experiments/exp_024_ltx2_prompt_sweep/config.yaml'

with open(CONFIG_PATH) as f:
    cfg = yaml.safe_load(f)

prompts      = {}
samples_meta = {}
for s in cfg['samples']:
    sid = s['sample_id']
    prompts[sid]      = {v['category']: v['text'] for v in s['prompt_variants']}
    samples_meta[sid] = s.get('class', '')

CATEGORIES = ['A_empty', 'A_word', 'B', 'C', 'D', 'E']
CAT_LABELS = {
    'A_empty': 'A — Empty string',
    'A_word' : 'A — "transition"',
    'B'      : 'B — Generic cinematic',
    'C'      : 'C — Abstract descriptor',
    'D'      : 'D — Typed creative',
    'E'      : 'E — Timed',
}

def show_sample(sample_id):
    cls   = samples_meta.get(sample_id, '')
    parts = sample_id.split('__')
    start = parts[1].replace('-',' ').title() if len(parts) > 1 else ''
    end   = parts[2].replace('-',' ').title() if len(parts) > 2 else ''
    display(Markdown(f'---\n## {start} → {end}\n*{cls}*'))
    for cat in CATEGORIES:
        cat_dir = RUN_DIR / sample_id / cat
        finals  = sorted(p for p in cat_dir.glob('*.mp4') if 'stage1' not in p.name)
        if not finals:
            continue
        prompt = (prompts.get(sample_id) or {}).get(cat, '(empty)')
        display(Markdown(f'**{CAT_LABELS[cat]}**'))
        display(Video(str(finals[0]), embed=False, width=600))
        display(Markdown(f'<details><summary>Prompt</summary>{prompt}</details>'))

print('✓ Ready — run show_sample() cells below')
```
